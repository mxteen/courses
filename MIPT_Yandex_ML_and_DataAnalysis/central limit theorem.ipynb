```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import exp
from math import pi
```

```python
#Генерация выборки из экспоненциального распределения
l=1
sample=np.random.exponential(l, size=1000)
#Построение гистограммы по сгенерированной выборке (рекомендуемый в задании параметр Normed устарел)
plt.hist(sample,bins=30,density = True,label='fraction of samples')
plt.ylabel('f(x)')
plt.xlabel('$x$')
#Построение теоретической плотности распределения
x= np.linspace(0, 10, 100)
pdf= np.linspace(0, 10, 100)
i=0
for element in x:
    pdf[i] = l*exp(-l*element)
    i+=1
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
```

```python
#создание выборок объема 5,10 и 50
i=0
sample5=np.zeros((1000,5))
sample10=np.zeros((1000,10))
sample50=np.zeros((1000,50))
ave5=np.arange(1000)
ave10=np.arange(1000)
ave50=np.arange(1000)
for i in range(1000):
    sample5[i] = np.random.choice(sample, 5)
    sample10[i] = np.random.choice(sample, 10)
    sample50[i] = np.random.choice(sample, 50)
    i+=1
#вычисление выборочный средних
ave5=np.mean(sample5, axis = 1)
ave10=np.mean(sample10, axis = 1)
ave50=np.mean(sample50, axis = 1)
```

```python
#построение гистограмм выборочных средних
plt.hist(ave5,label='fraction of samples 5')
plt.ylabel('DSM')
plt.xlabel('$x$')
plt.hist(ave10,label='fraction of samples 10')
plt.hist(ave50,label='fraction of samples 50')
plt.legend()
```

```python
#вычисление математического ожидания и дисперсии экспоненциального распределения
E=1/l
sig=1/(l**2)
# параметры нормального распределения:
mu=E
sig_norm5=sig/5
sig_norm10=sig/10
sig_norm50=sig/50
```

```python
#аппроксимация нормальным распределением выборки размера n=5
plt.hist(ave5,density = True,label='fraction of samples 5')
plt.ylabel('f(x)')
plt.xlabel('$x$')
i=0
norm5= np.linspace(0, 10, 100)
for element in x:
    norm5[i] = 1/((sig_norm5*2*pi)**0.5)*exp(-((element-mu)**2)/(2*sig_norm5))
    i+=1
plt.plot(x, norm5, label='normal pdf (n=5)')
plt.legend()
```

```python
#аппроксимация нормальным распределением выборки размера n=10
plt.hist(ave10,density = True,label='fraction of samples 10')
plt.ylabel('f(x)')
plt.xlabel('$x$')
i=0
norm10= np.linspace(0, 10, 100)
for element in x:
    norm10[i] = 1/((sig_norm10*2*pi)**0.5)*exp(-((element-mu)**2)/(2*sig_norm10))
    i+=1
plt.plot(x, norm10, label='normal pdf (n=10)')
plt.legend()
```

```python
#аппроксимация нормальным распределением выборки размера n=50
plt.hist(ave50,density = True,label='fraction of samples 50')
plt.ylabel('f(x)')
plt.xlabel('$x$')
i=0
norm50= np.linspace(0, 10, 100)
for element in x:
    norm50[i] = 1/((sig_norm50*2*pi)**0.5)*exp(-((element-mu)**2)/(2*sig_norm50))
    i+=1
plt.plot(x, norm50, label='normal pdf (n=50)')
plt.legend()
```

# Вывод 
При возрастании параметра n увеличивается число выборочных средних, значения которых близки к математическому ожиданию. Точность аппроксимации распределения выборочных средних нормальным распределением с ростом n возрастает, в том числе в области, близкой к нулю, где исхордное экспоненциальное распределение имеет абсолютный максимум.
