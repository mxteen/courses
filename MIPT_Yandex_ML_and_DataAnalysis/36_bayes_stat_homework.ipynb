```python
from sklearn.datasets import load_breast_cancer, load_digits
from sklearn.model_selection import cross_val_score, GridSearchCV #, train_test_split
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
import pandas as pd
```

```python
def write_answer(value, filename):
    """
    Функция для записи ответов.
    * value - значение, которое нужно записать в файл
    * filename - имя файла
    
    """
    with open(filename+".txt", "w") as file:
        file.write(str(value))
```

**Вопрос 1** Каким получилось максимальное качество классификации на датасете breast_cancer?

```python
# Загрузка данных
data, target = load_breast_cancer(return_X_y=True, as_frame=True)
```

```python
data.head()
```

```python
# создание объекток классификаторов
classifier_BernoulliNB = BernoulliNB()
classifier_MultinomialNB = MultinomialNB()
classifier_GaussianNB = GaussianNB()

# список объектов классификаторов
classifiers = [classifier_BernoulliNB, classifier_MultinomialNB, classifier_GaussianNB]
mean_scores = []

# Сравниваем работу классификаторов
for classifier in classifiers:
    scores = cross_val_score(estimator=classifier, X=data, y=target, cv=5)
    mean_scores.append(scores.mean())
    print(classifier, scores.mean(), scores)
    
write_answer(mean_scores[2], 'bayes_stat_homework_1')
```

Верно. В датасете breast_cancer признаки вещественные, и лучшее качество получилось у GaussianNB

**Вопрос 2** Каким получилось максимальное качество классификации на датасете digits?

```python
data.head()
```

```python
# создание объекток классификаторов
classifier_BernoulliNB = BernoulliNB()
classifier_MultinomialNB = MultinomialNB()
classifier_GaussianNB = GaussianNB()

# список объектов классификаторов
classifiers = [classifier_BernoulliNB, classifier_MultinomialNB, classifier_GaussianNB]
mean_scores = []

# Сравниваем работу классификаторов
for classifier in classifiers:
    scores = cross_val_score(estimator=classifier, X=data, y=target, cv=5)
    mean_scores.append(scores.mean())
    print(classifier, scores.mean(), scores)
    
write_answer(mean_scores[1], 'bayes_stat_homework_2')
```
