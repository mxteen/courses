## Уменьшение размерности

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

from collections import  Counter
```

```python
# попробуем провести эксперименты только с первой тысячкей объектов
data = digits.data[:1000]
labels = digits.target[:1000]
```

```python
# посмотрим, что уполучилась сбалансированная выборка,
# то есть, что все объекты встречаются приблизительно
# в одинаковой пропорции, объекты всех классов.
# Сделаем это с помощью Counter.

print(Counter(labels))
```

```python
pylab.figure(figsize = (10, 6))
pylab.bar(Counter(labels).keys(), Counter(labels).values())
```

Построим некоторый классификатор. Выбираем классификатор K ближайших соседей.
И давайте сделаем следующее: обучим его на всей выборке, на всех тех 1000 цифрах, которые мы выбрали, и посмотрим, какое получается качество. Тык мы поймём, насколько наши объекты легко классифицировать по их изображениям.


```python
classifier = KNeighborsClassifier()
```

```python
classifier.fit(data, labels)
```

```python
print(classification_report(classifier.predict(data), labels))
```

Теперь давайте сделаем следующее: если мы перейдем в пространство меньшей размерности, то есть перейдем в пространство 2, нам снова будет интересно, остается ли это условие в силе. То есть в новом пространстве наши цифры так же находятся близко друг к другу или уже нет, или уже в ближайшем окружении начнутся встречаться объекты чужих классов, и тогда наше качество классификации уменьшится.

### Random projection

```python
from sklearn import random_projection
```

```python
# Строим объект SparseRandomProjection и передаем ему следующие аргументы.
# Первый аргумент — это количество новых компонент, которые нас интересуют
projection = random_projection.SparseRandomProjection(n_components = 2, random_state = 0)


# После того, как мы построили наш объект, нужно это отображение обучить на исходных данных и применить,
# то есть получить данные в новой размерности.

# Сделаем это с помощью метода fit_predict. 
data_2d_rp = projection.fit_transform(data)
```

```python
pylab.figure(figsize=(10, 6))
pylab.scatter(data_2d_rp[:, 0], data_2d_rp[:, 1], c = labels)
```

```python
# оцениваем качество на обучающей выборке

classifier.fit(data_2d_rp, labels)
print(classification_report(classifier.predict(data_2d_rp), labels))
```

**Вывод**. Качество в среднем упало до 50 %, то есть мы видим, что для половины объектов ближайшие 5 соседей уже могут быть числами из разных классов. 

### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2, random_state = 0)
data_2d_pca = pca.fit_transform(data)
```

```python
pylab.figure(figsize = (10, 6))
pylab.scatter(data_2d_pca[:, 0], data_2d_pca[:, 1], c = labels)
```

```python
classifier.fit(data_2d_pca, labels)
print(classification_report(classifier.predict(data_2d_pca), labels))
```

### MDS

```python
from sklearn import manifold
```

```python
mds = manifold.MDS(n_components = 2, n_init = 1, max_iter = 100)
data_2d_mds = mds.fit_transform(data)
```

```python
pylab.figure(figsize=(10, 6))
pylab.scatter(data_2d_mds[:, 0], data_2d_mds[:, 1], c = labels)
```

```python
classifier.fit(data_2d_mds, labels)
print(classification_report(classifier.predict(data_2d_mds), labels))
```

### t- SNE

```python
tsne = manifold.TSNE(n_components = 2, init = 'pca', random_state = 0)
data_2d_tsne = tsne.fit_transform(data)
```

```python
pylab.figure(figsize = (10, 6))
pylab.scatter(data_2d_tsne[:, 0], data_2d_tsne[:, 1], c = labels)
```

```python
classifier.fit(data_2d_tsne, labels)
print(classification_report(classifier.predict(data_2d_tsne), labels))
```
