## Эксперты курса

При выполнении заданий у вас могут возникнуть трудности. В таком случае вы всегда можете обратиться к команде экспертов и задать свой вопрос в Discord.

* Команда экспертов помогает по вопросам курса и не гарантирует, что даст ответ на вопросы, не связанные с обучением. 
* При общении с коллегами и командой поддержки просим вас соблюдать правила вежливого общения. Не грубите и не используйте нецензурную лексику. Соблюдение этих правил является частью оферты и обязательно при прохождении курса. При нарушении мы имеем право закрыть доступ к курсу.
* Эксперты отвечают на вопросы по заданиям курса в общих каналах. Команда экспертов не работает в личных сообщениях, а отвечает на вопросы в ветках. Это обусловлено двумя основными причинами:
    - Команда экспертов работает сменами, у членов команды есть определённые рабочие часы и выходные. Написав в лс кому-то из команды экспертов, вы можете попасть на нерабочее время конкретного эксперта + по окончании смены ваш вопрос не сможет перехватить другой человек.
    - Вопросы часто дублируются, поэтому важно обсуждать все открыто, чтобы другие студенты со схожей проблемой имели возможность прочитать обсуждение и, в случае чего, сразу разобраться с проблемой, не задавая повторяющийся вопрос снова.
* Призвать экспертов можно специальным тегом @`expert`.

### Cервер

Сервер — это просто особенно мощный компьютер, которым, как правило, пользуется больше одного человека одновременно. Он предоставляет доступ к данным и инструментам другим ПК

Что нужно знать о сервере?
* Гораздо мощнее обычного ПК ( несколько десятков-сотен ядер и сотни-тысячи гигабайт оперативной памяти)
* Необходим для задач, требующих значительной вычислительной мощности (тренировки крупных моделей машинного обучения, прогоны симуляций, проверки математических гипотез, анализ астрофизических данных, сборки геномов или любых других данных, которые больше оперативной памяти обычного компьютера)
* Все суперкомпьютеры (или почти все) это сервера или кластеры (объединение компьютеров в одну вычислительную систему). Исходно они создавались для того, чтобы можно было шарить и эффективно распределять ресурсы. Поэтому там нет графического интерфейса (он потребляет ресурсы, которые можно потратить на вычисления)
* Для работы на сервере используют три вида ОС: Windows, UNIX и UNIX-подобные ОС (в частности, Linux). Доминирующая часть ОС на серверах это Linux — более 95%, около 2% используют Unix и Windows
* Файлы, находящиеся на вашем ПК, никак не связаны с сервером. Поэтому если вы хотите открыть какой-то файл с вашего компьютера, его прежде нужно загрузить на сервер.

## Расписание 

### Подробнее про расписание

1.  Уроки открываются **3 раза в неделю**: в понедельник, среду и пятницу. Открытие происходит в 19:00 по московскому времени.
2.  На решение одного урока отводится 2 недели. После наступает дедлайн. В течение двух недель за задание можно получить **полный балл**. После дедлайна — **половину баллов**. Обратите внимание, что в задачах с рецензированием дедлайны работают иначе — если вы не успели отправить решение до дедлайна, то сможете отправить ссылку только один раз в течение недели. Через неделю после дедлайна возможность послать работу закрывается. Следите за расписанием, чтобы не упустить сроки. Дедлайны наступают в 19:00 по московскому времени. 
3.  После дедлайна открывается доступ к разбору урока. Иногда там разбираются альтернативные варианты решения и темы, которых не было на лекции. Если вам не удалось решить задачу, вы сможете посмотреть вариант ее решения, а если вы справились, то сможете сравнить свое решение с решением от преподавателя.

## Работа с кодом

Поскольку вы будете писать много кода с первых лекций, рассказываем, как избежать проблем при его написании, а главное, как его понятно, красиво и правильно оформлять 🙂😉

**Комментировать свой код**

* Это очень важно.
* Так вы никогда не забудете, что делает тот или иной блок кода, и всегда сможете быстро что-то поменять, сохраняя общую логику.
* Это служит и дополнительной проверкой работоспособности вашего кода, так как вы всегда будете уверены, что вы делаете и зачем.
* Также, если вы отправите код с комментариями в Discord, то команде поддержки будет гораздо проще оперативно выявить причину, почему ничего не работает.
* Старайтесь комментировать код всегда!

**Тестировать код в JupyterHub**
* Всегда прогоняйте свой код в JupyterHub перед отправкой решения на LMS.
* Это тоже очень важно.
* Без проверки кода на адекватность существует очень высокий риск, что где-то вы допустили ошибку. От такого не застрахован никто, всегда легко опечататься и написать неправильно переменную, забыть отступ или создать бесконечный цикл. Поэтому лишний раз перепроверить себя не повредит.

**Сохраняйте все решения**
* Начните сразу собирать репозиторий со своими решениями заданий для портфолио.
* Если вы хотите добавить условия задачи из курса к себе в репозиторий, пожалуйста, укажите авторство Karpov.Courses и оставьте ссылку на [страницу](https://karpov.courses/simulator) курса. 
* Попробуйте добавлять в репозиторий не только задания из курса, но и собственные пет-проекты. Это поможет вам впечатлить будущего работодателя. 
* Пусть ваши решения по всем заданиям будут где-то сохранены. Из них потом можно будет создать себе неплохое портфолио для будущих работодателей. Если не знаете, как оформить репозиторий, то можете почитать эту [статью](https://t.me/KarpovCourses/212).
* Пример правильно оформленного кода с комментариями вы можете посмотреть [тут](https://disk.yandex.ru/d/cztp-Kur3ZhgCw).
